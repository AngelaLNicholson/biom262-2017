# Parsing Log.final.out files from STAR

Here are the instructions to parse the Log.final.out files generated by STAR to quickly generate summary statistics for all of the samples that you have aligned in one batch. This parser requires that you have all of the files in one directory (each one with its own unique prefix).

This was written by Olga Botvinnik, a bioinformatics graduate student in the Yeo lab. She has a lot of resources available on github that she has written and packaged up nicely to share with the community. The github page for her file parsers can be found [here](https://github.com/olgabot/rnaseek).

I recommend making a software directory in your home folder to keep all the code you download from Github. 

    mkdir ~/software
    
Move into that directory and clone the github repo there:

    cd software
    
The command to clone a repo is git clone and you copy the link from the big green button on the github page. Click the button to get the download URL. 

    git clone https://github.com/olgabot/rnaseek.git
    
This will make a directory in that folder called rnaseek. Follow the install instructions on the github page (scroll down they are at the bottom).

    cd rnaseek
    pip install .
    
Take a look at the documentation to see what you can learn about how the parsers work. I think the install will automatically make the script available in your anaconda bin. Notice there are multiple parsers written for outputs of various programs. Here is an example of how I would run the one for STAR

    cd /path/to/output/that/contains/Log.final.out/fils
    python ~/anaconda2/bin/combine_star_mapping_stats.py 
    
Take a look at the parameters that this function requires shown under the parser_add_argument [section](https://github.com/olgabot/rnaseek/blob/master/rnaseek/scripts/combine_star_mapping_stats.py). Why does running this with just the command work? What does my output look like? 

Use SSHFS to easily move the .csv output file to your desktop and view it in excel. Or if you want to get fancy, load it up in R studio.
