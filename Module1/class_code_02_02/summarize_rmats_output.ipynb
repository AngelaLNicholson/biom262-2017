```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
directory = "/projects/ps-yeolab/biom262_2017/rmats/"
SE = pd.read_table(directory+"U2AF1-LV08-K562-SE.MATS.JunctionCountOnly.txt",
                  index_col=0)
SE.head()
```

```python
SE.columns
```

```python
columns_to_looks_at = [u'GeneID', u'geneSymbol', u'chr', u'strand', u'exonStart_0base',
       u'exonEnd',u'IJC_SAMPLE_1', u'SJC_SAMPLE_1',
       u'IJC_SAMPLE_2', u'SJC_SAMPLE_2',u'PValue',
        u'FDR', u'IncLevel1', u'IncLevel2', u'IncLevelDifference']
```

```python
print SE.shape
SE[columns_to_looks_at].head()
```

```python
sns.boxplot(SE['FDR'])
```

```python
sns.boxplot(SE['IncLevelDifference'])
```

```python
#FDR < 0.05, and abs(IncLevelDifference) > 0.2

SE_filtered = SE.loc[(SE["FDR"] < 0.05) & 
                    (abs(SE['IncLevelDifference']) > 0.2)]
print SE.shape
print SE_filtered.shape
SE_filtered.head()
```

```python
sns.boxplot(SE_filtered['IncLevelDifference'])
```

```python
plt.hist(SE_filtered['IncLevelDifference'], bins=100)
```

```python
SE_filtered.sort(columns = 'IncLevelDifference', ascending=False)[columns_to_looks_at].head(50)
```
